```python
from src.models.crnn.crnn import run_crnn
formatted_results = run_crnn(create_data=False, retrain=False)
```

```python
from src.utils.generate_ground_truth import generate_gt_events_dict

GT_EVENT_DICT = generate_gt_events_dict()
```

```python
from collections import Counter

total = 0.0
matched = 0.0

for filename, ground_truth_event in GT_EVENT_DICT.items():
    total += 1.0
    detected_events = formatted_results.get(filename, [])
    if len(detected_events) == 0:
        continue
    all_labels = [detected_event['event_label'] for detected_event in detected_events]
    label_counts = Counter(all_labels)
    most_frequent_label = label_counts.most_common(1)[0][0]
    if ground_truth_event[0]['event_label'] == most_frequent_label:
        matched += 1.0

matched / total
```
