```python
DATA_ROOT_PATH = "/Users/yixian/Library/CloudStorage/OneDrive-GeorgiaInstituteofTechnology/Courses/25_3_Fall/CS7643_DL/sound_event_detection/data/raw/us8k/"
FOLD_MAPPING = {
    "1": "fold1",
    "2": "fold2",
    "3": "fold3",
    "4": "fold4",
    "5": "fold5",
    "6": "fold6",
    "7": "fold7",
    "8": "fold8",
    "9": "fold9",
    "10": "fold10",
}
OUTPUT_PATH = "/Users/yixian/Library/CloudStorage/OneDrive-GeorgiaInstituteofTechnology/Courses/25_3_Fall/CS7643_DL/sound_event_detection/data/raw/us8k/audio_classes/"

def extract_us8k_class(class_name, output_path, metadata_df):
    class_df = metadata_df[metadata_df['class'] == class_name]
    os.makedirs(output_path + f"/{class_name}", exist_ok=True)
    for index, row in class_df.iterrows():
        fold_id = row['fold']
        audio_path = os.path.join(DATA_ROOT_PATH, FOLD_MAPPING[str(fold_id)], row['slice_file_name'])
        shutil.copy(audio_path, os.path.join(output_path + f"/{class_name}", row['slice_file_name']))

for class_name in metadata_df['class'].unique():
    extract_us8k_class(class_name, OUTPUT_PATH, metadata_df)
```
